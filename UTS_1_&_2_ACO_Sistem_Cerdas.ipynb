# MENGUTAMAKAN JARAK TERDEKAT = WAKTU TERSINGKAT

```python
import numpy as np

def euclidean_distance(a, b):
    return np.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

locations = {
    'A': (15, 5, 'Red'),
    'B': (10, 30, 'Green'),
    'C': (5, 10, 'Green'),
    'D': (15, 20, 'Green'),
    'E': (25, 10, 'Green'),
    'F': (30, 20, 'Yellow'),
    'G': (45, 5, 'Green'),
    'H': (35, 30, 'Green'),
    'I': (45, 25, 'Red'),
    'J': (40, 10, 'Green')  # Starting point
}

# Initialize pheromone levels and heuristic
pheromones = {src: {dest: 0.1 for dest in locations if src != dest} for src in locations}
heuristic = {src: {dest: 1 / euclidean_distance(locations[src][:2], locations[dest][:2]) for dest in locations if src != dest} for src in locations}
initial_pheromone = 0.1
evaporation_rate = 0.1
pheromone_deposit = 0.5  # Increased deposit rate for better routes

distances = {src: {dest: euclidean_distance(locations[src][:2], locations[dest][:2]) for dest in locations if src != dest} for src in locations}

n_ants = 10
n_iterations = 100
best_route = None
best_distance = float('inf')

for iteration in range(n_iterations):
    for ant in range(n_ants):
        current = 'J'
        route = [current]
        visited = set(route)
        # Generate a route starting at 'J'
        while len(route) < len(locations):
            possible_cities = set(locations.keys()) - visited
            next_city = max(possible_cities, key=lambda x: pheromones[current][x] ** 2 * heuristic[current][x])
            route.append(next_city)
            visited.add(next_city)
            current = next_city

        route_distance = sum(distances[route[i]][route[i+1]] for i in range(len(route) - 1))

        # Update the best route found
        if route_distance < best_distance:
            best_distance = route_distance
            best_route = route

    # Update pheromones for all routes
    for src in pheromones:
        for dest in pheromones[src]:
            pheromones[src][dest] *= (1 - evaporation_rate)  # Evaporation

    # Strengthen pheromones on the best route
    for i in range(len(best_route) - 1):
        pheromones[best_route[i]][best_route[i + 1]] += pheromone_deposit / distances[best_route[i]][best_route[i + 1]]

    print(f"Iteration {iteration + 1}: Best route so far: {best_route}, Distance: {best_distance} KM")

print("Final best route:", best_route)
print("Final distance:", best_distance, "KM")
```

# MENGUTAMAKAN KEUNTUNGAN

```python
import numpy as np

def euclidean_distance(a, b):
    return np.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

def calculate_cost_and_time(route, distances):
    total_distance = 0
    total_time = 0
    total_cost = 50000  # Biaya awal

    for i in range(len(route) - 1):
        distance = distances[route[i]][route[i + 1]]
        total_distance += distance
        total_time += distance  # Kecepatan 1 KM/Menit

        # Tambahkan biaya bahan bakar
        total_cost += distance * 1000

    return total_cost, total_distance, total_time

def evaluate_profit(route, distances, locations):
    cost, distance, time = calculate_cost_and_time(route, distances)
    time_elapsed = 0
    for loc in route:
        if locations[loc][2] == 'Yellow' and time_elapsed > 120:
            cost += (time_elapsed - 120) // 30 * 5000
        elif locations[loc][2] == 'Red':
            if time_elapsed <= 180:
                cost -= 25000
            else:
                cost += (time_elapsed - 180) // 30 * 10000
        time_elapsed += distances[route[route.index(loc) - 1]][loc] if route.index(loc) != 0 else 0

        # Tambahkan bonus untuk setiap pengiriman berhasil
        cost -= 15000

    # Keuntungan adalah negatif dari biaya karena kita mengurangi biaya dari pendapatan
    profit = -cost
    return profit, distance

locations = {
    'A': (15, 5, 'Red'),
    'B': (10, 30, 'Green'),
    'C': (5, 10, 'Green'),
    'D': (15, 20, 'Green'),
    'E': (25, 10, 'Green'),
    'F': (30, 20, 'Yellow'),
    'G': (45, 5, 'Green'),
    'H': (35, 30, 'Green'),
    'I': (45, 25, 'Red'),
    'J': (40, 10, 'Green')
}

# Initialize pheromone levels
pheromones = {src: {dest: 0.1 for dest in locations if src != dest} for src in locations}
distances = {src: {dest: euclidean_distance(locations[src][:2], locations[dest][:2]) for dest in locations if src != dest} for src in locations}
n_ants = 10
n_iterations = 100
best_route = None
max_profit = float('-inf')

for iteration in range(n_iterations):
    for ant in range(n_ants):
        route = ['J'] + np.random.permutation([loc for loc in locations if loc != 'J']).tolist()
        profit, distance = evaluate_profit(route, distances, locations)

        if profit > max_profit:
            max_profit = profit
            best_route = route

    print(f"Iteration {iteration + 1}: Best route so far: {best_route}, Profit: {max_profit}, Distance: {distance} KM")

# Final output
print("Final best route:", best_route)
print("Final profit:", max_profit)
```
